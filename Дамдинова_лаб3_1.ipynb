```python
#№ 3.1.1 Напишите функцию, которая принимает строку и возвращает её в
#перевернутом виде, не используя циклы.
def reverse_string(s):
    """Вернуть перевернутую строку 's'.

    Параметры:
        - s (str): исходная строка.

    Результат:
        - str: перевернутая строка."""
    return s[::-1]


print(reverse_string("можно зачет"))
```

```python
#№ 3.1.2 Напишите функцию, которая извлекает доменное имя из адреса
#электронной почты.
def extract_domain(email):
    """Извлечь доменное имя из адреса электронной почты 'email'.

    Параметры:
        - email (str): адрес электронной почты.

    Результат:
        - str: доменное имя."""
    # Находим позицию символа @
    at_position = email.find('@')

    # Если символ @ не найден, возвращаем сообщение об ошибке
    if at_position == -1:
        return "Ошибка: некорректный email адрес"

    # Извлекаем часть строки после символа @
    domain = email[at_position + 1:]
    return domain


# Примеры использования
email1 = "user.name@example.com"
email2 = "test@company.org"
email3 = "invalid-email"

print(f"Домен из '{email1}': {extract_domain(email1)}")
print(f"Домен из '{email2}': {extract_domain(email2)}")
print(f"Домен из '{email3}': {extract_domain(email3)}")
```

```python
# № 3.1.3 Напишите функцию для подсчета количества слов в предложении. Cлова
#разделены одним или несколькими пробелами. Знаки препинания не учитывать.
def count_words(sentence):
    """Подсчитать количество слов в предложении 'sentence'.

    Параметры:
        - sentence (str): предложение.

    Результат:
        - int: количество слов."""
    # Удаляем лишние пробелы и разбиваем на слова
    words = sentence.strip().split()
    return len(words)


text = "Дайте зачет пожалуйста"
print(count_words(text))
```

```python
#№ 3.1.4 Напишите функцию, которая принимает список и возвращает новый
#список, содержащий только уникальные элементы из исходного
def get_unique_elements(input_list):
    """Вернуть список с уникальными элементами из 'input_list'."""
    unique_list = []
    for item in input_list:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list

my_list = [1, 2, 3, 2, 4, 3, 5, 1]
print(get_unique_elements(my_list))
```

```python
#№ 3.1.5 Напишите функцию, которая проверяет, является ли строка палиндромом
#(читается одинаково в обоих направлениях), игнорируя регистр и пробелы.
def is_palindrome(s):
    """Проверить, является ли строка 's' палиндромом.

    Параметры:
        - s (str): строка для проверки.

    Результат:
        - bool: True, если палиндром, иначе False."""
    # Убираем пробелы и приводим к нижнему регистру
    cleaned = s.replace(" ", "").lower()
    # Сравниваем строку с её перевернутой версией
    return cleaned == cleaned[::-1]


print(is_palindrome("А роза упала на лапу Азора"))
print(is_palindrome("Привет мир"))
```

```python
#№ 3.1.6 В компании принят формат кода товара: три буквы, дефис, три цифры,
#дефис, две буквы (например, XYZ-123-AB). Напишите функцию, которая
#форматирует строку вида xyz123ab в нужный формат.
def format_product_code(code):
    """Отформатировать код товара 'code'.

    Параметры:
        - code (str): исходный код в виде сплошной строки.

    Результат:
        - str: отформатированный код."""
    part1 = code[:3].upper()
    part2 = code[3:6]
    part3 = code[6:].upper()
    return f"{part1}-{part2}-{part3}"


print(format_product_code("xyz123ab"))
```

```python
#№ 3.1.7 Напишите функцию, которая принимает список чисел и возвращает их
#среднее арифметическое.
def calculate_average(numbers):
    """Вычислить среднее арифметическое для списка 'numbers'.

    Параметры:
        - numbers (list): список чисел.

    Результат:
        - float: среднее арифметическое."""
    if not numbers:
        return 0.0  # Возвращаем 0 для пустого списка
    return sum(numbers) / len(numbers)


my_numbers = [10, 20, 30, 40, 50]
print(f"{calculate_average(my_numbers):.2f}")
```

```python
#№ 3.1.8 Напишите функцию, которая принимает текст и список "стоп-слов".
#Функция должна заменять все стоп-слова в тексте на символ *.
def censor_text(text, stop_words):
    """Заменить стоп-слова в тексте 'text' на '*'.

    Параметры:
        - text (str): исходный текст.
        - stop_words (list): список стоп-слов.

    Результат:
        - str: обработанный текст."""
    words = text.split()
    result = []
    for word in words:
        if word.lower().strip('.,!?;:') in stop_words:
            result.append('*')
        else:
            result.append(word)
    return ' '.join(result)


text = "Этот плохой пример содержит плохие слова."
stop_list = ["плохой", "плохие"]
print(censor_text(text, stop_list))
```

```python
#№ 3.1.9 Напишите функцию для генерации корпоративного email-адреса. Она
#должна принимать имя и фамилию, преобразовывать их в
#формат i.ivanov@company.com.
def generate_corporate_email(first_name, last_name):
    """Сгенерировать email из имени 'first_name' и фамилии 'last_name'.

    Параметры:
        - first_name (str): имя пользователя.
        - last_name (str): фамилия пользователя.

    Результат:
        - str: корпоративный email."""
    first_letter = first_name[0].lower()
    last_lower = last_name.lower()
    return f"{first_letter}.{last_lower}@mgpu.ru"


print(generate_corporate_email("kristina", "damdinova"))
```

```python
#№ 3.1.10 Напишите функцию, которая принимает список строк и возвращает
#новый список, содержащий только те строки, длина которых больше 5 символов.
def filter_long_strings(strings):
    """Отфильтровать строки из списка 'strings' по длине (> 5).

    Параметры:
        - strings (list): список строк.

    Результат:
        - list: отфильтрованный список."""
    result = []
    for word in strings:
        if len(word) > 5:
            result.append(word)
    return result


word_list = ["аналитика", "данные", "отчет", "бизнес", "план", "модель"]
print(filter_long_strings(word_list))
```

```python
#№ 3.1.11 Напишите функцию, которая создает акроним (аббревиатуру) из фразы.
#Например, из "Key Performance Indicator" должно получиться "KPI".
def create_acronym(phrase):
    """Создать акроним из фразы 'phrase'.

    Параметры:
        - phrase (str): исходная фраза.

    Результат:
        - str: акроним в верхнем регистре."""
    words = phrase.split()
    acronym = ""
    for word in words:
        acronym += word[0].upper()
    return acronym


print(create_acronym("Key Performance Indicator"))
print(create_acronym("business process modeling notation"))
```

```python
#№ 3.1.12 Напишите функцию для форматирования телефонного номера. Она
#должна принимать строку из 11 цифр (например, 89211234567) и возвращать ее в
#формате +7 (921) 123-45-67.
def format_phone_number(phone):
    """Отформатировать телефонный номер 'phone'.

    Параметры:
        - phone (str): строка из 11 цифр.

    Результат:
        - str: отформатированный номер."""
    return f"+7 ({phone[1:4]}) {phone[4:7]}-{phone[7:9]}-{phone[9:]}"


print(format_phone_number("89211234567"))
```

```python
#№ 3.1.13 Напишите функцию, которая разбирает строку в формате CSV (CommaSeparated Values) и возвращает список ее полей.
def parse_csv_row(csv_row):
    """Разобрать строку 'csv_row' в формате CSV.

    Параметры:
        - csv_row (str): строка с данными, разделенными запятыми.

    Результат:
        - list: список полей из строки."""
    return csv_row.split(',')


row = "Иванов,Иван,35,Менеджер"
print(parse_csv_row(row))
```

```python
#№ 3.1.14 Напишите функцию, которая находит самое длинное слово в списке
#слов.
def find_longest_word(words):
    """Найти самое длинное слово в списке 'words'.

    Параметры:
        - words (list): список слов.

    Результат:
        - str: самое длинное слово."""
    longest = ""
    for word in words:
        if len(word) > len(longest):
            longest = word
    return longest


word_list = ["аналитика", "данные", "отчет", "бизнес", "модель", "прогнозирование"]
print(find_longest_word(word_list))
```

```python
#№ 3.1.15 В целях безопасности необходимо маскировать номера кредитных карт.
#Напишите функцию, которая принимает номер карты (строку из 16 цифр) и
#возвращает его в виде, где первые 12 цифр заменены на *.
def mask_card_number(card_number):
    """Замаскировать номер карты 'card_number'.

    Параметры:
        - card_number (str): номер карты (16 цифр).

    Результат:
        - str: замаскированный номер."""
    return card_number[:4] + " **** **** " + card_number[-4:]


print(mask_card_number("4567890123456789"))
```

```python
#№ 3.1.16 Напишите функцию, которая сортирует список полных имен ("Имя
#Фамилия") по фамилии.
def sort_by_last_name(names):
    """Отсортировать список 'names' по фамилии.

    Параметры:
        - names (list): список полных имен.

    Результат:
        - list: отсортированный список."""
    return sorted(names, key=lambda name: name.split()[1])


name_list = ["Иван Петров", "Анна Сидорова", "Петр Иванов", "Мария Алексеева"]
print(sort_by_last_name(name_list))
```

```python
#№ 3.1.17 Напишите функцию, которая "сплющивает" вложенный список (список
#списков) в один плоский список.
def flatten_list(nested_list):
    """Преобразовать вложенный список 'nested_list' в плоский.

    Параметры:
        - nested_list (list): список, содержащий другие списки.

    Результат:
        - list: плоский список."""
    result = []
    for sublist in nested_list:
        for item in sublist:
            result.append(item)
    return result


matrix = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
print(flatten_list(matrix))
```

```python
#№ 3.1.18 Напишите функцию, которая удаляет все знаки препинания из текста.
def remove_punctuation(text):
    """Удалить все знаки препинания из 'text'.

    Параметры:
        - text (str): исходный текст.

    Результат:
        - str: текст без знаков препинания."""
    result = ""
    for char in text:
        if char not in ".,!?;:-()\"'":
            result += char
    return result


input_text = "Привет, мир! Как твои дела, Python?"
print(remove_punctuation(input_text))
```

```python
#№ 3.1.19 Напишите функцию, которая объединяет два списка и возвращает новый
#отсортированный список без дубликатов.
def merge_and_sort_unique(list1, list2):
    """Объединить два списка 'list1' и 'list2', отсортировать и убрать дубликаты.

    Параметры:
        - list1 (list): первый список.
        - list2 (list): второй список.

    Результат:
        - list: объединенный, отсортированный, уникальный список."""
    combined = list1 + list2
    unique = list(set(combined))
    return sorted(unique)


list_a = [3, 1, 5, 2, 5]
list_b = [8, 2, 4, 1, 6]
print(merge_and_sort_unique(list_a, list_b))
```

```python
#№ 3.1.20 Напишите функцию, которая извлекает все хэштеги (слова,
#начинающиеся с #) из строки.
def extract_hashtags(text):
    """Извлечь хэштеги из текста 'text'.

    Параметры:
        - text (str): текст с хэштегами.

    Результат:
        - list: список хэштегов без символа '#'."""
    hashtags = []
    words = text.split()
    for word in words:
        if word.startswith('#'):
            hashtags.append(word[1:])
    return hashtags


post = "Сегодня отличный день! #python #программирование #бизнесинформатика"
print(extract_hashtags(post))
```
